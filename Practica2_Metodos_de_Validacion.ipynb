Práctica 2 Métodos de Validación
I. Utilizando el dataset metodosDeValidacion.csv elabore un programa en python que realice lo siguiente:
    • Sin mezclar los datos genere un conjunto de entrenamiento (60%) y uno de prueba (40%).
    • Con validación cruzada genere conjuntos de validación para k=3.
    • Genere conjuntos de validación con Leave One Out
    • Utilizando Bootstrap genere 2 conjuntos de validación con conjuntos de entrenamiento de tamaño=5

II.El programa deberá imprimir en pantalla los conjuntos de validación
generados por cada método de validación

```python
#Importar liberias
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.model_selection import LeaveOneOut
from sklearn.model_selection import KFold
from sklearn.utils import resample
```

```python
# Funcion para la validación cruzada
def validacionCruzada(X_train,y_train,k):
	print ('\n ----------------------\n')
	print ('\n Validación cruzada \n')
	kf = KFold(n_splits=k)
	print ('\n X \n')
	for train, test in kf.split(X_train):
		print(X_train[train], X_train[test])
	print ('\n y \n')
	for train, test in kf.split(X_train):
		print('train',y_train[train], 'test' ,y_train[test])

# Funcion LeaveOneOut
def dejaUnoFuera(X_train,y_train):
	print ('\n ----------------------\n')
	print ('\n LeaveOneOut')
	loo = LeaveOneOut()
	print (loo.get_n_splits(X_train))
	print ('\n X \n')
	for i, (train_index, test_index) in enumerate(loo.split(X_train)):
		print('\n',i,X_train[train_index], X_train[test_index])
	print ('\n y \n')
	for i, (train_index, test_index) in enumerate(loo.split(X_train)):
		print('\n',i,'train',y_train[train_index], 'test' ,y_train[test_index])

#Funcion para el bootstrap
def bootstrap(X_train,y_train,nc,m):
	print ('\n ----------------------\n')
	print ('\n Bootstrap')
	print ('\n X \n')
	for i in range(nc):
		train = resample(X_train, n_samples = m)
		test = np.array([x for x in X_train if x not in train])
		print ('train',train,'test',test)
	print ('\n y')
	for i in range(nc):
		train = resample(y_train, n_samples = m)
		test = np.array([x for x in y_train if x not in train])
		print ('train',train,'test',test)
	print ('\n')
```

```python
df = pd.read_csv('metodosDeValidacion.csv',sep=',')
print(df)
```

```python
X = df[['x']].values
y = df['y'].values
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, shuffle=False)
print ('\n X_train \n',X_train)
print ('\n X_test \n',X_test)
print ('\n y_train \n',y_train)
print ('\n y_test \n',y_test)
```

```python
validacionCruzada(X_train,y_train,3)
```

```python
dejaUnoFuera(X_train,y_train)
```

```python
bootstrap(X_train,y_train,2,5)
```
