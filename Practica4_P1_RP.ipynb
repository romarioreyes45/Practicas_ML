```python
#Importar liberias
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import SGDRegressor
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.discriminant_analysis import StandardScaler
import operator
import matplotlib.pyplot as plt
```

```python
#Visualizar dataset
df = pd.read_csv('datos.csv',sep=',')
print(df)
```

```python
#Divivdir el dataset en 70% entrenamiento y 30% test
x = df[['x']].values
y = df['y'].values
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, shuffle=True, random_state=0)
print(x_train)
print(x_test)
print(y_train)
print(y_test)
```

```python
#Regresion lineal con OLS
reg_1_OLS = LinearRegression()
reg_1_OLS.fit(x_train, y_train)
y_pred_1_OLS = reg_1_OLS.predict(x_test)
mse_1_OLS = mean_squared_error(y_test, y_pred_1_OLS)
r2_1_OLS = r2_score(y_test, y_pred_1_OLS)

#Graficar regresion lineal con OLS
plt.scatter(x_test, y_test, color='purple')
plt.plot(x_test, y_pred_1_OLS, color='green', linewidth=3)
plt.title('Regresion Lineal con OLS')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
#Regresion polinomial de grado 2 con OLS
reg_2_OLS = PolynomialFeatures(degree=2)
x_train_2_OLS = reg_2_OLS.fit_transform(x_train)
x_test_2_OLS = reg_2_OLS.transform(x_test)
regresion_2_OLS = LinearRegression()
regresion_2_OLS.fit(x_train_2_OLS, y_train)
y_pred_2_OLS = regresion_2_OLS.predict(x_test_2_OLS)
mse_2_ols = mean_squared_error(y_test, y_pred_2_OLS)
r2_2_ols = r2_score(y_test, y_pred_2_OLS)

#Graficar regresion polinomial con de grado 2 con OLS
plt.scatter(x_test, y_test, color='purple')
sort_axis = operator.itemgetter(0)
sorted_zip = sorted(zip(x_test, y_pred_2_OLS))
x_test2, y_pred_2_OLS = zip(*sorted_zip)
plt.plot(x_test2, y_pred_2_OLS, color='green', linewidth=3)
plt.title('Regresion Polinomial de grado 2 con OLS')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
#Regresion polinomial de grado 3 con OLS
reg_3_OLS = PolynomialFeatures(degree=3)
x_train_3_OLS = reg_3_OLS.fit_transform(x_train)
x_test_3_OLS = reg_3_OLS.transform(x_test)
regresion_3_OLS = LinearRegression()
regresion_3_OLS.fit(x_train_3_OLS, y_train)
y_pred_3_OLS = regresion_3_OLS.predict(x_test_3_OLS)
mse_3_ols = mean_squared_error(y_test, y_pred_3_OLS)
r2_3_ols = r2_score(y_test, y_pred_3_OLS)

#Graficar regresion polinomial con de grado 3 con OLS
plt.scatter(x_test, y_test, color='purple')
sort_axis = operator.itemgetter(0)
sorted_zip = sorted(zip(x_test, y_pred_3_OLS))
x_test3, y_pred_3_OLS = zip(*sorted_zip)
plt.plot(x_test3, y_pred_3_OLS, color='green', linewidth=3)
plt.title('Regresion Polinomial de grado 3 con OLS')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
#Regresion lineal con SGD
reg_1_SGD = SGDRegressor(learning_rate='constant', eta0=0.0000001, max_iter=10000)
reg_1_SGD.fit(x_train, y_train)
y_pred_1_SGD = reg_1_SGD.predict(x_test)
mse_1_SGD = mean_squared_error(y_test, y_pred_1_SGD)
r2_1_SGD = r2_score(y_test, y_pred_1_SGD)

#Graficar regresion lineal con SGD
plt.scatter(x_test, y_test, color='purple')
plt.plot(x_test, y_pred_1_SGD, color='green', linewidth=3)
plt.title('Regresion Lineal con SGD')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
#Regresion polinomial de grado 2 con SGD
reg_2_SGD = PolynomialFeatures(degree=2, include_bias=False)
x_train_2_SGD = reg_2_SGD.fit_transform(x_train)
x_test_2_SGD = reg_2_SGD.transform(x_test)
regresion_2_SGD = SGDRegressor(learning_rate='constant', eta0=0.0000001, max_iter=10000)
regresion_2_SGD.fit(x_train_2_SGD, y_train)
y_pred_2_SGD = regresion_2_SGD.predict(x_test_2_SGD)
mse_2_sgd = mean_squared_error(y_test, y_pred_2_SGD)
r2_2_sgd = r2_score(y_test, y_pred_2_SGD)


#Graficar regresion polinomial con de grado 2 con SGD
plt.scatter(x_test, y_test, color='purple')
sort_axis = operator.itemgetter(0)
sorted_zip = sorted(zip(x_test, y_pred_2_SGD))
x_test2, y_pred_2_SGD = zip(*sorted_zip)
plt.plot(x_test2, y_pred_2_SGD, color='green', linewidth=3)
plt.title('Regresion Polinomial de grado 2 con SGD')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
#Regresion polinomial de grado 3 con SGD
reg_3_SGD = PolynomialFeatures(degree=3)
x_train_3_SGD = reg_3_SGD.fit_transform(x_train)
x_test_3_SGD = reg_3_SGD.transform(x_test)
regresion_3_SGD = SGDRegressor(learning_rate='constant', eta0=0.0000001, max_iter=10000)
regresion_3_SGD.fit(x_train_3_SGD, y_train)
y_pred_3_SGD = regresion_3_SGD.predict(x_test_3_SGD)
mse_3_sgd = mean_squared_error(y_test, y_pred_3_SGD)
r2_3_sgd = r2_score(y_test, y_pred_3_SGD)


#Graficar regresion polinomial con de grado 3 con SGD
plt.scatter(x_test, y_test, color='purple')
sort_axis = operator.itemgetter(0)
sorted_zip = sorted(zip(x_test, y_pred_3_SGD), key=sort_axis)
x_test3, y_pred_3_SGD = zip(*sorted_zip)
plt.plot(x_test3, y_pred_3_SGD, color='green', linewidth=3)
plt.title('Regresion Polinomial de grado 3 con SGD')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
#Resumen de resultados
print('Regresion lineal con OLS')
print('mse: ', mse_1_OLS)
print('r2: ', r2_1_OLS)
print('Regresion polinomial de grado 2 con OLS')
print('mse: ', mse_2_ols)
print('r2: ', r2_2_ols)
print('Regresion polinomial de grado 3 con OLS')
print('mse: ', mse_3_ols)
print('r2: ', r2_3_ols)
print('Regresion lineal con SGD')
print('mse: ', mse_1_SGD)
print('r2: ', r2_1_SGD)
print('Regresion polinomial de grado 2 con SGD')
print('mse: ', mse_2_sgd)
print('r2: ', r2_2_sgd)
print('Regresion polinomial de grado 3 con SGD')
print('mse: ', mse_3_sgd)
print('r2: ', r2_3_sgd)
```
